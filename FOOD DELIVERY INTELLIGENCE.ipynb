```python

import pandas as pd
data=pd.read_csv('Swiggy_dataset.csv.zip')
data
```

```python
data.drop('uuid',axis=1,inplace=True)
data.drop('id',axis=1,inplace=True)
data.drop('address',axis=1,inplace=True)
data
```

```python
import numpy as np
data=data.replace('--',np.nan)
data
```

```python
data.rename(columns={'totalRatingsString':'total rating'},inplace=True)
data
```

```python
data.isnull().any()
```

```python
data.dtypes
```

```python
data['avgRating']=data['avgRating'].astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.kdeplot(data['avgRating'])
plt.xlabel('avgRating')
plt.ylabel('Density')
plt.title('Average Rating Distribution')
plt.show()
```

```python
data['avgRating']=data['avgRating'].fillna(data['avgRating'].median())
data
```

```python
data['locality']=data['locality'].fillna('Unknown')
data['unserviceable']=data['unserviceable'].fillna('False')
data['veg']=data['veg'].fillna('False')
data['City']=data['city'].fillna('Unknown')
data
```

```python
data.isnull().any()
```

```python
data['costoftwo']=data['costForTwoStrings'].str.extract('(\d+)').astype(float)
```

```python

import matplotlib.pyplot as plt
plt.boxplot(data['costoftwo'])
plt.ylabel('costoftwo')
plt.title('boxplot of cost')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(data['avgRating'])
plt.ylabel('avgRating')
plt.title('boxplot of Average Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(data['deliveryTime'])
plt.ylabel('delivaryTime')
plt.title('boxplot of Delivery Time')
plt.show()
```

```python
data.columns=data.columns.str.lower()
data.columns=data.columns.str.strip()
data=data.loc[:,~data.columns.duplicated()]
```

```python

data['cuisines']=data['cuisines'].str.replace("' '","', '")
```

```python
data['cuisines']
```

```python
import ast
data['cuisines']=data['cuisines'].apply(ast.literal_eval)
```

```python

data['cuisines']=data['cuisines'].apply(tuple)
```

```python
data.drop_duplicates(inplace=True)
```

```python
#1 TOP 10 FAVORITE CUISINES-PIE CHART
cuisine_counts = data['cuisines'].explode().value_counts()
top10 = cuisine_counts.head(10)
plt.figure(figsize=(6,7))
plt.pie(
    top10.values,
    labels=top10.index,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor':'white'},
    pctdistance=0.85
)
plt.set_cmap('viridis')
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Top 10 Cuisines on Swiggy by Restaurant Count", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#2 Distribution of cost-HISTOGRAM
plt.figure(figsize=(10,6))
plt.hist(data['costoftwo'],bins=30,edgecolor='black',color='navy',alpha=0.9)
plt.title('Distribution Of Cost',fontsize=16,fontweight='bold')
plt.xlabel('cost for two(INR)',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#3 average rating distribution-histogram
plt.figure(figsize=(10,6))
plt.hist(data['avgrating'],bins=20,edgecolor='black',color='navy',alpha=0.9)
plt.title('Distribution of Average Rating',fontsize=16,fontweight='bold')
plt.xlabel('Rating',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#4 veg vs non-veg restaurants-countplot
plt.figure(figsize=(8,5))
ax=sns.countplot(data=data, x='veg',edgecolor='black',color='navy',alpha=0.9)
plt.title('Veg VS Non Veg Restaurants',fontsize=16,fontweight='bold')
plt.xlabel('Resataurant Type',fontsize=12)
plt.ylabel('No.of Restaurants',fontsize=12)
ax.set_xticklabels(['Non-Veg','Veg'])
for container in ax.containers:
  ax.bar_label(container,padding=3)
plt.show()
```

```python
#5 CITY WISE RESTAURANT COUNT-BAR PLOT
plt.figure(figsize=(10,6))
city_counts = data['city'].value_counts().head(10)
ax = sns.barplot(
    x=city_counts.values,
    y=city_counts.index,
    edgecolor='black',
    color='navy',
    alpha=0.9
)
plt.title("Top 10 Cities with the Most Restaurants", fontsize=16, fontweight='bold')
plt.xlabel("Number of Restaurants", fontsize=12)
plt.ylabel("City", fontsize=12)
for i, v in enumerate(city_counts.values):
    plt.text(v + 1, i, str(v), va='center', fontsize=11)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
#6 DELIVARY TIME DISTRIBUTION-HISTOGRAM
plt.hist(data['deliverytime'],bins=20,color='navy')
plt.title('Delivery Time Distribution',fontsize=16,fontweight='bold')
plt.xlabel(' Delivery Time(mins)',fontsize=12)
plt.ylabel('No.of orders / Resataurants',fontsize=12)
plt.show()
```

```python
#7 relationship between cost and rating-SCATTER PLOT
plt.scatter(data['costoftwo'],data['avgrating'],alpha=0.6)
plt.title('COST VS RATING')
plt.xlabel('cost')
plt.ylabel('Rating')
plt.show()
```

```python
#8 DISTRIBUTION OF TOATAL RATING CATEGORIES-BAR CHART
rating_order=data['total rating'].value_counts().index
plt.figure(figsize=(10,5))
sns.countplot(data=data,y='total rating',order=rating_order,color='navy')
plt.title('Total Rating Category Distribution',fontsize=16,fontweight='bold')
plt.xlabel('Count')
plt.ylabel('Rating Category')
plt.show()
```

```python
#9 Top localities with most restaurants
top_localities=data['locality'].value_counts().head(10)
plt.figure(figsize=(10,5))
ax=sns.barplot(x=top_localities.values,y=top_localities.index,color='navy',alpha=0.9)
plt.title('Top 10 Localities With Most Restaurants',fontsize=16,fontweight='bold')
plt.xlabel('No.of restaurants',fontsize=12)
plt.ylabel('Locality',fontsize=12)
for container in ax.containers:
  ax.bar_label(container)
plt.show()
```

```python
#10 heatmap of feature relationships
plt.figure(figsize=(8,5))
sns.heatmap(data[['avgrating','costoftwo','deliverytime']].corr(),annot=True,fmt='.2f',linewidths=0.5,square=True,cbar_kws={'shrink':.8})
plt.title('Correlation between key variables',fontsize=16,fontweight='bold')
plt.show()
```
